```python
missing_values = countries_data.isnull().sum()
print(missing_values)
```

```python
happiest_country = countries_data.loc[countries_data['Happiness Score'].idxmax()]
print(f"The happiest country is: {happiest_country['Country']}")
```

```python
happiest_country_name = happiest_country['Country']

for column in ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity', 'Dystopia Residual']:
  top_country_in_column = countries_data.loc[countries_data[column].idxmax()]
  if top_country_in_column['Country'] == happiest_country_name:
    print(f"The happiest country, {happiest_country_name}, is also the top country in {column}.")
else:
  print(f"The happiest country, {happiest_country_name}, is not the top country in any of the specified")
```

```python
country = 'Jamaica'
happiness_score = countries_data.loc[countries_data['Country'] == country, 'Happiness Score'].values[0]
print(f"The Happiness Score for {country} is: {happiness_score}")
```

```python
top_10_happiest = countries_data.sort_values('Happiness Score', ascending=False).head(10)
print(top_10_happiest[['Country', 'Happiness Score']])
```

```python
import matplotlib.pyplot as plt

correlation_matrix = countries_data.select_dtypes(include=['number']).corr()
print(correlation_matrix)

plt.figure(figsize=(12, 10))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar()
plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

The correlation matrix shows strong positive relationships between happiness scores and factors like Economy (GDP per Capita), Family, Health (Life Expectancy), Freedom, Trust (Government Corruption), and Generosity. These correlations indicate that economic stability, social support, and good health significantly contribute to happiness. Additionally, weaker correlations with Dystopia Residual suggest that some factors affecting happiness remain unexplained.

```python
region_generosity = countries_data.groupby('Region')['Generosity'].mean()
most_generous_region = region_generosity.idxmax()
print(f"The region with the highest average generosity is: {most_generous_region}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
import pandas as pd
import numpy as np

data1 = pd.Series(np.random.normal(0, 1, 100))
data2 = pd.Series(np.random.normal(1, 1.5, 100))
```

```python
statistic, p_value = stats.ks_2samp(data1, data2)
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(np.sort(data1), np.sort(data2))
plt.plot([np.min(data1), np.max(data1)], [np.min(data2), np.max(data2)], color='red', linestyle='--')
plt.xlabel('Data 1 Quantiles')
plt.ylabel('Data 2 Quantiles')
plt.title('Q-Q Plot of Data 1 vs. Data 2')
```

What?: This Q-Q plot compares the quantiles of two datasets (Data 1 and Data 2). When the points align closely with the red dashed line, it suggests that the datasets share similar distributions. Any deviations from the line indicate variations in the shapes of the distributions.

Why?: Q-Q plots are helpful for checking if two datasets have similar distributions and for spotting any outliers or differences in their tails. In this case, the plot helps us see if Data 1 and Data 2 come from the same distribution, which is important for the Kolmogorov-Smirnov test.

```python
print(f"K-S statistic: {statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```
