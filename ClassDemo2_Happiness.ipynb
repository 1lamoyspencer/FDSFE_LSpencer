```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
rng=np.random.RandomState(1415)
```

```python
x= stats.norm.rvs(loc=0,scale=1,size=50,random_state=rng)
plt.hist(x, density=True)
plt.xlabel("Random Variable X")
plt.ylabel("Probability ")
plt.title("Gussian Distribution with Mean O and Standard Deviation 1")

t=np.linspace(-5,5,num=100)
pdf= stats.norm.pdf(t, loc=0, scale=1)

plt.plot(t,pdf)

plt.show()
```
